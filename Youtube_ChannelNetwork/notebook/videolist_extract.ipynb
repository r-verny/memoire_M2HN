```python
### On veut extraire la liste de toutes les vidéos étudiées
import os
from tqdm import tqdm
import pandas as pd
```

# Extraction de la liste de video id

```python
path = os.path.join(os.path.dirname(os.path.realpath("../../Youtube_SentimentAnalysis/data/videolist_filtered")),"videolist_filtered")
```

```python
id_list = []
for i, fichier_videolist in tqdm(enumerate(os.listdir(path))):
    df_videolist = pd.read_csv(os.path.join(path,fichier_videolist), sep= "\t", encoding= "utf-8")
    list_temp = df_videolist["videoId"].tolist()
    id_list.append(list_temp)
```

```python
id_list_final = [x for l in id_list for x in l]
```
